```python
big_df = stop_times.merge(trips, on='trip_id').merge(routes, on='route_id').merge(stops, on='stop_id').merge(calendar, on='service_id')
```

```python
rt8 = big_df[(big_df['route_id'] == '8') & (big_df['start_date'] < 20240701) & (big_df['end_date'] > 20240701) & (big_df['tuesday'] == 1) & (big_df['direction_id'] == 0)].sort_values('arrival_time')
rt8
```
